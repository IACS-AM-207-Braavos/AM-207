```python
import pandas as pd
import glob
```

```python
bayes_fs = glob.glob('mnist_semi_bayes/mnist_semi_bayes_2/*.json')
```

```python
ml_fs = glob.glob('mnist_semi_ml/mnist_semi_ml_2/*.json')
```

```python
bayes_acc_semi=[]
bayes_acc_sup=[]
for f in bayes_fs:
    df = pd.read_json(f, lines=True)
    bayes_acc_semi.append(df.semi_supervised_acc)
    bayes_acc_sup.append(df.supervised_acc)
```

```python
ml_acc_semi=[]
ml_acc_sup=[]
for f in ml_fs:
    df = pd.read_json(f, lines=True)
    ml_acc_semi.append(df.semi_supervised_acc)
    ml_acc_sup.append(df.supervised_acc)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_vals = np.linspace(0, 10000, len(bayes_acc_semi))
plt.figure(figsize=(10,7))
plt.plot(x_vals, bayes_acc_sup, label='bayes', alpha=1)
plt.plot(x_vals, ml_acc_sup, '-o', label='ml', alpha=0.3)
plt.title('Bayes vs ML Gans: MNIST Semi-Supervised Accuracy (N=1000)')
plt.xlabel('iteration')
plt.ylabel('accuracy')
plt.grid()
plt.legend()
plt.show()
```
